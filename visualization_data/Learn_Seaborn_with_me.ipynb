```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names()) # helps to get the inbuilt datasets present in Seaborn and use one of then for practice
```

```python
df = sns.load_dataset("diamonds")
df.head()
```

```python
df.shape
```

```python
df["cut"].value_counts()
```

```python
df.columns
```

```python
sns.scatterplot(x="carat",y="price",data=df, hue="cut", hue_order=[
"Premium","Very Good","Ideal","Good","Fair"])
plt.show()
```

```python
palette_color={"Premium":"blue","Very Good":"green","Ideal":"yellow","Good":"orange","Fair":"red"}
sns.countplot(x="cut", data=df, hue="cut",palette=palette_color)
plt.show()
```

```python
sns.set_theme(style='darkgrid',font='sans-serif',palette="Spectral")
sns.relplot(x="carat", y="price", 
            data=df,
            kind="scatter", 
            col="cut",
            row="color")

# Show plot
plt.show()
```

```python
price_12k_15k_df = df[(df["price"] > 12000) & (df["price"] < 15000)]
```

```python
# Create a palette from a Matplotlib colormap
gradient_palette = sns.color_palette("viridis", n_colors=df["clarity"].nunique())
```

```python
sns.set_theme(style='white',font='sans-serif',palette="light:#a15e9c")
#sns.set_theme(style='white',font='sans-serif',palette="colorblind")
sns.relplot(x="carat", y="price", data=price_12k_15k_df, kind="line",errorbar=None,style="clarity",markers=True,palette=gradient_palette,hue="clarity")
plt.show()
```

# categorical variable - view

```python
sns.set_theme()
gradient_palette = sns.color_palette("viridis", n_colors=df["clarity"].nunique())
sns.catplot(x="clarity",data=df,kind="count",palette=gradient_palette,hue="clarity")
plt.show()
```

```python
sns.catplot(x="clarity",y="price",data=df,kind="bar")
plt.show()
```
