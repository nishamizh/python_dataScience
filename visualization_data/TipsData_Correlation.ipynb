```python
#Import Python packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Supress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
for cols in data.columns:
  if data[cols].dtype=='category':
    print(cols, ' : ' , data[cols].unique())
```

```python
data.info()
```

```python
data.isnull().mean()*100
```

```python
sns.heatmap(data.corr(numeric_only=True),
            annot=True)
```

```python
sns.scatterplot(data=data,
                x='total_bill',
                y='tip',
                color='purple')
```

```python
sns.pairplot(data=data,
                x_vars=['total_bill','size'],
                y_vars=['tip'],
                kind='reg',
                height=5, aspect=1)

plt.show()
```

```python
# Understanding the TIP with (Total_bill,head_count) and tip based on Category of sex
sns.pairplot(data=data,
             x_vars=['total_bill','size'],
             y_vars=['tip'],
             hue='sex',
             kind='reg', height=5, aspect=1)

plt.show()
```

```python
# Understanding the TIP with (Total_bill,head_count) and tip based on Category of time

sns.pairplot(data=data,
             x_vars=['total_bill','size'],
             y_vars=['tip'],
             hue='time',
             kind='reg', height=5, aspect=1)

plt.show()
```

```python
# Understanding the TIP with (Total_bill,head_count) and tip based on Category of day

sns.pairplot(data=data,
             x_vars=['total_bill','size'],
             y_vars=['tip'],
             hue='day',
             kind='reg', height=5, aspect =2)

plt.show()
```

**Conclusion:**
The Tips is compared with Total Bill and Head Count for a Table under various categories such as Gender, Time and Day at which the table played the tip and found that The Tip is closely related to the Total Bill and Head Count of the table always.
